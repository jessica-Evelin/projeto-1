```python
# interagindo com usuario
num1 = input('digitar primeiro numero:')
num2 = input('digitar sugundo numero:')


#converção
num1_convert = int(num1)
num2_convert = int(num2)

#operação
operacao = input('escolha uma operação ( + , - , * , / , // , **):')
if operacao == '+':
  resultado =  (num1_convert + num2_convert)
  print('o resultado da adiçõa é :', resultado)
elif operacao == '-':
  resultado =  (num1_convert - num2_convert)
  print('o resultado da subtração é:', resultado)
elif operacao == '*':
  resultado =  (num1_convert * num2_convert)
  print('o resultado da multiplicação é:', resultado)
elif operacao == '/':
  resultado = (num1_convert / num2_convert)
  print('o resultado da divição é:', resultado)
elif operacao == '//':
  resultado =  (num1_convert // num2_convert)
  print('o resultado da divisão inteira é;', resultado)
elif operacao == '**':
  resultado =  (num1_convert  ** num2_convert)
  print('o resultado da potencia é:', resultado)
else:
  print('operação invalida!')
```
